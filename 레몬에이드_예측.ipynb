**01)과거의 데이터를 준비합니다!**

```python
#표를 만드는 판다스 라이브러리 가져오기
import pandas as pd
```

```python
#파일들로 부터 데이터 읽어오기
파일경로= 'https://raw.githubusercontent.com/rkqlthgms/Study_Python/main/lemonade.csv'
레모네이드= pd.read_csv(파일경로)
```

```python
#변수안에 어떤데이터가 있는지 확인 (데이터 모양으로 확인)
print(레모네이드.shape)
```

```python
#칼럼 이름 출력
print(레모네이드.columns)
```

```python
독립=레모네이드[['온도']]
종속=레모네이드[['판매량']]
print(독립.shape, 종속.shape)
```

```python
#실제데이터 출력
레모네이드.head()
```

```python
import tensorflow as tf
```

```python
#모델구조 만들기
X=tf.keras.layers.Input(shape=[1])
Y=tf.keras.layers.Dense(1)(X)
model=tf.keras.models.Model(X,Y)
model.compile(loss='mse')
```

```python
#모델을 몇번 학습시킬 것인가
model.fit(독립, 종속, epochs=600)
```

```python
#모델 예측
print("Predictions:", model.predict([[15]]))
```
